```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Hough(object):
    def __init__(self):
        self.image = cv2.imread('myimage.jpg')
        self.image_copy = np.copy(self.image)
        self.image_copy = cv2.cvtColor(self.image_copy, cv2.COLOR_BGR2RGB)
        self.gray = cv2.cvtColor(self.image_copy, cv2.COLOR_RGB2GRAY)
        self.edges = cv2.Canny(self.gray, 120, 240)
        self.minLineLength = 100
        self.lineGap = 10
        self.lines = cv2.HoughLinesP(self.edges, 1, np.pi/180, 100, self.minLineLength, self.lineGap)
        for x1, y1, x2, y2 in self.lines[1]:
            cv2.line(self.image_copy, (x1, y1), (x2, y2), (0, 0, 255), 3)

        f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 12))
        ax1.set_title("Normal")
        ax1.imshow(self.image_copy, cmap="gray")
        ax2.set_title("Filtered")
        ax2.imshow(self.edges, cmap="gray")
        plt.show()
        print(self.image.shape)
```

```python
Hough()
```
